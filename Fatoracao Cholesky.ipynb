```python
import numpy as np
from numpy.linalg import inv, det
import math
from sympy import symbols, solve
```

```python
#Entrada da matriz
A = np.array([[1,2,-3,0,3],[2,5,-1,1,4],[-3,-1,50,1,-19],[0,1,1,6,0],[3,4,-19,0,39]])
B = np.array([17,41,-45,30,51])
```

```python
def metCholesky(A,B):
  j=0
  autho = 1

  # Testa se pode ser decomposta
  while j < len(A):
    n=0
    while n < len(A):
      if A[j,n] != A[n,j]:
        autho = 0
      n = n + 1
    j = j + 1
  
  #Testa determinantes
  if autho == 1:
    j=1
    while j < len(A)+1:
      C = np.zeros((j,j), dtype='f')
      
      i=0
      while i < j:
        n=0
        while n < j: 
            C[i,n] = A[i,n]
            n = n + 1
        i = i + 1
      if(det(C)<0):
        autho = 0
      j = j + 1

    i=0
    G = np.zeros((len(A),len(A)),dtype='f')

    while i < len(A):
      j=0
      while j < len(A):
        if i == 0 and j == 0:
          G[i,j] = math.sqrt(A[i,j])
        elif i == j:
          k = 0
          aux=0
          while k < i:
            aux = aux + math.pow(G[i][k],2) 
            k = k + 1
          G[i,j] = math.sqrt(A[i,j] - aux)
        elif j==0 and j != i:
          G[i,j] = A[i,j]/G[j,j]
        elif j != i and j<i:
          aux=0
          k=0
          while k < j:
            aux = G[i][k]*G[j][k]
            k = k + 1
          G[i,j] = (A[i,j] - aux)/G[j,j]

        j = j + 1
      i = i + 1
    
    G_T = np.zeros((len(A),len(A)),dtype='f')
    i=0
    while i < len(A):
      j=0
      while j<len(A):
        G_T[i,j] = G[j,i]
        j = j + 1
      i = i + 1
    #print("G: \n", G)
    #print("G_T: \n", G_T)
    #print("B: ", B)
    Y =  np.linalg.solve(G,B)
    print("Y: ", Y)
    X = np.linalg.solve(G_T,Y)
    print("X: ", X)

  else:
    print("[Nao pode ser fatorada por Cholesky ]")

  
  
```

```python
metCholesky(A,B)
```
