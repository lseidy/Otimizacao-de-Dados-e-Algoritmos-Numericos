```python
import math
import numpy as np
```

```python
def metRK4(f,h,a,b,yo,m):
  xt=a; yt=yo
  vet_x = np.zeros(m)
  vet_y = np.zeros(m)
  vet_x[0] = xt
  vet_y[0] = yt

  for i in range(1,m):
    x = xt ; y = yt; k1 = f(x,y)
    x = xt+h/2; y = yt + (h/2)*k1; k2 = f(x,y)
    y = yt + (h/2)*k2; k3 = f(x,y)
    x = xt +h; y= yt + h*k3; k4 = f(x,y)
    xt = a + i*h; yt = yt+(h/6)*(k1+2*(k2+k3)+k4)
    vet_x[i] = xt; vet_y[i] = yt
    print("k1: ", k1, "\nk2: ", k2, "\nk3: ", k3, "\nk4: ",k4)

  print(vet_x)
  print(vet_y)
```

```python
f_x = lambda x,y: -(x*y)
metRK4(f_x,0.1,0,1,1,10)
```
