```python
import math
import numpy as np
from sympy import *
```

```python
vet_x = np.array([-1,0,2])
fx_y = np.array([4,1,-1])
```

```python
def metLagrange(vet_X, fx_Y, value, p):
  n=0
  p_n=0
  pn_print=""
  x = Symbol('x')
  for i in vet_X:
    t_mult = 1
    tdiv_mult = 1

    for j in vet_X:
      #Soma valores para divisao Lk
      if i != j:
        t_mult = t_mult * (x-j)
        tdiv_mult = tdiv_mult * (i-j)



    # Soma P_n = Yi*Lk
    p_n = p_n + fx_Y[n]*t_mult/tdiv_mult
    n=n+1

  print("Pn(x): ", p_n)
  print("Valor interpolado para ", value, ": ", round(p_n.subs(x,value),5))
```

```python
metLagrange(vet_x,fx_y,1,5)
```
