```python
def metFakePosition(f, a, b, i, p):
  if(f(a)*f(b) < 0):
    
    if (b-a < p):
      x = (b+a)/2
      return
    else:
      k = 1

    while True:
      
      x = (a*f(b)-b*f(a))/(f(b)-f(a))
      
      if (f(a)*f(b) < 0):
        b = x
      else:
        a = x

      k = k+1

      if (math.sqrt(f(x)**2) < p):
        break
      elif (k-1) == i:
        break
    print("n Interacoes:", k-1) 
    print("valor de x:", round(x,len(str(p))-2))   
```
