```python
import numpy as np
from numpy.linalg import inv, det
import math
from sympy import symbols, solve
```

```python
#Entrada da matriz
#A = np.array([[10,2,1],[1,5,1],[2,3,10]])
A = np.array([[10,2,1],[1,5,1],[2,3,10]])
B = np.array([7,-8,6]) 
```

```python
def metGaussSeidel(A , b, p , MAX):
  
  X = np.zeros(len(A))
  V = np.zeros(len(A))
  i=0
  Iter =0
  
  #Calcula X0
  while i < len(A):
    X[i] = b[i]/A[i][i]
    i = i+1

  while True:  
    Iter= Iter + 1
    Num = 0
    NormaDen = 0
    i=0
    
    while i < len(A):
      Sum = 0
      j = 0

      # reliza soma necessaria para calcular Xk
      while j < len(A):
        if i != j:
          Sum = Sum + A[i][j]*X[j]
        j = j + 1
      
      #Faz o calculo de Xk e armazena em X
      V[i] = X[i]
      X[i] = (b[i]-Sum)/A[i][i]
      t = abs(V[i]-X[i])
      
      if t > Num:
        Num = t
      if abs(X[i] > NormaDen):
        NormaDen = abs(X[i])
      i = i + 1
    
    NormaRel = Num/NormaDen
    if NormaRel <= p or Iter >= MAX:
      break
  if NormaRel <= p:
    print("Convergiu")
    print("X: ", X)
    print("Iter: ", Iter)
  else:
    print("Nao convergiu")
```

```python
metGaussSeidel(A,B,0.05,5)
```
