```python
def metNewtonRaphson(f ,f_line, x, i, p):
  k=1
  if abs(f(x)) < p:
    return

  while True:
      #print("x: ", x)
      #print("f(x): ", f(x))
      #print("f_line(x)", f_line(x))
      #print("resultado: ", f(x)/f_line(x))
      #print("resultado x - fx: ", x - f(x))
      #print("resultado final: ", x - f(x)/f_line(x))
      #print("-----------------------------\n")

      x1 = x - f(x)/f_line(x)
    
      if (abs(f(x)) < p or k == i):
        #print("Modulo: ", abs(f(x)))
        break
      else:
        x = x1
      k = k +1

  print("interacoes: ", k)
  print("valor de x: ", round(x,5))
```
