**Aula de Método de Bissecção**



```python
def metBissec(f, a, b, i, p):
  if(f(a)*f(b) < 0):
    if (b-a < p):
      
      return (b+a)/2
    else:
      k = 1     

      while True:
        x = (a+b)/2


        if(f(a)*f(x) > 0):
          a = x
        else:
          b = x

        k = k + 1
        

        if(k-1) == i:
           break
        elif math.sqrt(f(x)**2) < p:
          break

      print("n Interacoes:", k-1) 
      print("valor de x:", round(x,len(str(p))-2))        
  else:
    print("nao pode ser calculada pro bissec")

```
