### Gauss com Pivotamento

```python
import numpy as np
from numpy.linalg import inv
import math
from sympy import symbols, solve
```

```python
#Entrada da matriz
A = np.array([[1.5,5.4,3.3,10],[4.2,2.3,4.5,11.7],[2.7,5.7,7.8,8.9]])
B = np.array([[1.5,5.4,3.3,10], [4.2,2.3,4.5,11.7],[2.7,5.7,7.8,8.9]])
```

```python
def metGaussPivo(A,p):
  k = -1
  m = 0

  for i in A:
    n = 1
    aux = []
    k = k + 1
    
    #Faz o primeiro pivotamento para k=0
    if k == 0:
      while n < len(A):
        if abs(A[n,k]) > abs(A[n-1,k]):
          m = n
        n = n + 1

      n = 0
      while n < len(A) + 1:

        aux = np.append(aux,A[m,n])
        n = n + 1
      
      A = np.delete(A,(m), axis = 0)
      aux = np.vstack([aux,A])
      A = aux
      #--------------------------#
    else:
      #Aqui fazemos o Gauss inicial para cada linha
      n = 0
      if k > 0:
        m = A[k,0]/A[0,0]

        for j in i:
          A[k,n] = A[k,n] - m*A[0,n]
          n = n + 1
  
  # Realizado pivotamento para as demais linhas dps da primeira
  k=-1
  for i in A:
    n=0
    k=k+1

    for j in i:
    
      if k > 1:
        n = k
        while n < len(A):
          if abs(A[n,k]) > abs(A[n-1,k]):
            m = n
          n = n + 1
        
        n = 0
        p=0
        new_A=[]

        while n < len(A):
          aux=[]
          if n != m:
            p=0
            while p < len(A) + 1:
              aux = np.append(aux,A[n,p])
              p=p+1
            if n == 0:
              new_A = aux
            else:
              new_A = np.vstack([new_A,aux])
          else:
            p=0
            while p < len(A) + 1:
              aux = np.append(aux,A[m,p])
              p=p+1
            new_A = np.vstack([new_A,aux])
          n = n + 1

        n = 0
        A = new_A
  #Pivotamento final termina aqui

  # Iteração de ajuste para formar triangulo superior para linhas
  # que precisam ser tratadas mais de 1 vez  
        n=0
        m_1 = A[k,k-1]/A[k-1,k-1]

        for j in i:                              
          A[k,n] = A[k,n] - m_1*A[k-1,n]
          n = n + 1


  #Resolução do Sistema de Equações
  k = len(A)
  i=0
  B =[]
  new_A=[]

  while (i < k): 

    B = np.append(B, A[i, k])
    n=0
    aux=[]

    while(n < k):
      if i == 0:
        new_A = np.append(new_A,A[i,n])    
      else:
        aux = np.append(aux, A[i,n])
      n=n+1
    if i >0:
      new_A = np.vstack([new_A,aux])
    i = i+1
  
  X = inv(new_A)
  S = np.matmul(X,B)
  print("A: \n", A)
  print("\nSolucao:", S) 
```
